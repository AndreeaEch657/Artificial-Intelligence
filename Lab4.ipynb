```python
import numpy as np

w = np.array([1,-1,0,0.5])
x1 = np.array([1,-2,1.5,0])
x2 = np.array([1,-0.5,-2,-1.5])
x3 = np.array([0,1,-1,1.5])
l = [x1,x2,x3]
it = 0

print("Neuroni bipolari binari:")

for x in l:
    it+=1
    net = np.dot(w,x)
    print("Net",it,":", net)
    w = w + np.sign(net)*x
    print("Weight",it+1,":", w)
    
w1 = np.array([1,-1,0,0.5])
x1 = np.array([1,-2,1.5,0])
x2 = np.array([1,-0.5,-2,-1.5])
x3 = np.array([0,1,-1,1.5])   
l = [x1,x2,x3]
it = 0

print("\nFct de activare continua si bipolara:")
    
for x in l:
    it+=1
    net1 = np.dot(w1,x)
    fnet1 = (2/(1+np.exp(-1*net1)))-1
    print("Net",it,":", fnet1)
    w1 = w1 + (((2/(1+np.exp(-1*net1)))-1)*x)
    print("Weight",it,":", w1)
    
print("\nEx:7")
x1 = np.array([1,-2])
x2 = np.array([0,1])
x3 = np.array([2,3])
w = np.array([1,-1])
l = [x1,x2,x3]
it = 0

print("\nNeuroni bipolari binari:")

for x in l:
    it+=1
    net = np.dot(w,x)
    print("Net",it,":", net)
    w = w + np.sign(net)*x
    print("Weight",it+1,":", w)

x1 = np.array([1,-2])
x2 = np.array([0,1])
x3 = np.array([2,3])
w1 = np.array([1,-1])
l = [x1,x2,x3]
it = 0

print("\nFct de activare continua si bipolara:")

for x in l:
    it+=1
    net1 = np.dot(w1,x)
    fnet1 = (2/(1+np.exp(-1*net1)))-1
    print("Net",it,":", fnet1)
    w1 = w1 + (((2/(1+np.exp(-1*net1)))-1)*x)
    print("Weight",it,":", w1)
    





```
