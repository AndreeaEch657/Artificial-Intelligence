```python
import numpy as np
import math

x1 = [45,85]
x2 = [50,43]
x3 = [40,80]
x4 = [55,42]
x5 = [200,43]
x6 = [48,40]
x7 = [195,41]
x8 = [43,87]
x9 = [190,40]

w1 = [50,89]
w2 = [30,47]
w3 = [100,38]

l = [x1,x2,x3,x4,x5,x6,x7,x8,x9]
p = [w1,w2,w3]
c = 1
epochs = 100

def dist(x1,w1): 
    return math.sqrt((x1[0] - w1[0]) ** 2 + (x1[1] - w1[1]) ** 2)

for i in range(epochs):
    print("Epoca",i+1)
    for x in l:
        print("Pattern:",x)
        distante = [dist(x,w1),dist(x,w2),dist(x,w3)]
        minim = np.argmin(distante)
        p[minim] = p[minim] + (c*(np.array(x) - np.array(p[minim])))
        print("Prototip castigator", p[minim],"\n")
        
print("Lista prototipuri finale", p)
        

```
